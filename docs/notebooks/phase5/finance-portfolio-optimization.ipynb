# Finance Portfolio Optimization
Calls `POST /v1/finance/portfolio_optimization` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"expected_returns": [0.1, 0.2, 0.12], "covariance_matrix": [[0.05, 0.01, 0.02], [0.01, 0.06, 0.01], [0.02, 0.01, 0.04]], "budget": 2, "risk_factor": 0.5, "solver": "qaoa", "optimizer": {"name": "cobyla", "maxiter": 25}, "reps": 1, "shots": 512, "seed": 7}
response = httpx.post(f"{base_url}/finance/portfolio_optimization", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
