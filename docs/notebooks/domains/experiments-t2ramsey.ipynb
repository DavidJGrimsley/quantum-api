# T2 Ramsey Experiment
Calls `POST /v1/experiments/t2ramsey` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"qubits": [0], "delays": [0.000001, 0.000002, 0.000003, 0.000004, 0.000005], "osc_freq": 100000.0, "shots": 128, "seed": 7}
response = httpx.post(f"{base_url}/experiments/t2ramsey", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
