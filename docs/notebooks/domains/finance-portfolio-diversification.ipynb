# Finance Portfolio Diversification
Calls `POST /v1/finance/portfolio_diversification` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"similarity_matrix": [[1.0, 0.2, 0.3], [0.2, 1.0, 0.4], [0.3, 0.4, 1.0]], "num_clusters": 2, "asset_labels": ["ALPHA", "BETA", "GAMMA"], "solver": "exact", "optimizer": {"name": "cobyla", "maxiter": 25}, "reps": 1, "shots": 512, "seed": 7}
response = httpx.post(f"{base_url}/finance/portfolio_diversification", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
