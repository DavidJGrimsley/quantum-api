# State Tomography
Calls `POST /v1/experiments/state_tomography` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"circuit": {"num_qubits": 1, "operations": [{"gate": "h", "target": 0}]}, "shots": 512, "seed": 7, "target_statevector": [{"real": 0.70710678, "imag": 0.0}, {"real": 0.70710678, "imag": 0.0}]}
response = httpx.post(f"{base_url}/experiments/state_tomography", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
