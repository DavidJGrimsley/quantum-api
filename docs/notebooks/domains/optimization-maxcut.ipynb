# Optimization MaxCut
Calls `POST /v1/optimization/maxcut` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"num_nodes": 3, "edges": [{"source": 0, "target": 1, "weight": 1.5}, {"source": 1, "target": 2, "weight": 2.0}, {"source": 0, "target": 2, "weight": 0.5}], "solver": "qaoa", "reps": 1, "optimizer": {"name": "cobyla", "maxiter": 25}, "shots": 512, "seed": 7}
response = httpx.post(f"{base_url}/optimization/maxcut", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
