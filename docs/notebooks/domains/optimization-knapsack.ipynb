# Optimization Knapsack
Calls `POST /v1/optimization/knapsack` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"item_values": [3, 4, 5], "item_weights": [2, 3, 4], "capacity": 5, "solver": "exact", "reps": 1, "optimizer": {"name": "cobyla", "maxiter": 25}, "shots": 512, "seed": 7}
response = httpx.post(f"{base_url}/optimization/knapsack", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
