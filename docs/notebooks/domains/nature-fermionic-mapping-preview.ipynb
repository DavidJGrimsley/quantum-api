# Nature Fermionic Mapping Preview
Calls `POST /v1/nature/fermionic_mapping_preview` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"atoms": [{"symbol": "H", "x": 0.0, "y": 0.0, "z": 0.0}, {"symbol": "H", "x": 0.0, "y": 0.0, "z": 0.735}], "basis": "sto3g", "charge": 0, "spin": 0, "mapper": "jordan_wigner", "ansatz": {"type": "real_amplitudes", "reps": 1}, "optimizer": {"name": "cobyla", "maxiter": 25}, "seed": 7}
response = httpx.post(f"{base_url}/nature/fermionic_mapping_preview", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
