# Optimization TSP
Calls `POST /v1/optimization/tsp` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"distance_matrix": [[0.0, 10.0, 15.0, 20.0], [10.0, 0.0, 35.0, 25.0], [15.0, 35.0, 0.0, 30.0], [20.0, 25.0, 30.0, 0.0]], "solver": "exact", "reps": 1, "optimizer": {"name": "cobyla", "maxiter": 25}, "shots": 512, "seed": 7}
response = httpx.post(f"{base_url}/optimization/tsp", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
