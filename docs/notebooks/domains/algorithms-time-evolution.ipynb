# Time Evolution
Calls `POST /v1/algorithms/time_evolution` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"variant": "trotter_qrte", "hamiltonian": [{"pauli": "Z", "coefficient": 1.0}], "time": 0.5, "initial_state": {"num_qubits": 1, "operations": [{"gate": "h", "target": 0}]}, "num_timesteps": 2, "shots": 256, "seed": 7}
response = httpx.post(f"{base_url}/algorithms/time_evolution", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
