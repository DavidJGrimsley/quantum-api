# Amplitude Estimation
Calls `POST /v1/algorithms/amplitude_estimation` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"variant": "ae", "state_preparation": {"num_qubits": 1, "operations": [{"gate": "ry", "target": 0, "theta": 1.2}]}, "objective_qubits": [0], "num_eval_qubits": 2, "shots": 256, "seed": 7}
response = httpx.post(f"{base_url}/algorithms/amplitude_estimation", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
