# Phase Estimation
Calls `POST /v1/algorithms/phase_estimation` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"variant": "standard", "unitary": {"num_qubits": 1, "operations": [{"gate": "z", "target": 0}]}, "state_preparation": {"num_qubits": 1, "operations": [{"gate": "h", "target": 0}]}, "num_evaluation_qubits": 3, "shots": 256, "seed": 7}
response = httpx.post(f"{base_url}/algorithms/phase_estimation", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
