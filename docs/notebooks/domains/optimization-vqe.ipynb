# Optimization VQE
Calls `POST /v1/optimization/vqe` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"pauli_sum": [{"pauli": "ZI", "coefficient": 1.0}, {"pauli": "IZ", "coefficient": -0.5}, {"pauli": "XX", "coefficient": 0.2}], "ansatz": {"type": "real_amplitudes", "reps": 1}, "optimizer": {"name": "cobyla", "maxiter": 25}, "shots": 512, "seed": 7}
response = httpx.post(f"{base_url}/optimization/vqe", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
