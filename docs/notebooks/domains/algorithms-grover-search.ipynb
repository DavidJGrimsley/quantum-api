# Grover Search
Calls `POST /v1/algorithms/grover_search` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"marked_bitstrings": ["11"], "iterations": [1], "sample_from_iterations": False, "shots": 256, "seed": 7}
response = httpx.post(f"{base_url}/algorithms/grover_search", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
