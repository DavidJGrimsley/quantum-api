# Quantum Volume
Calls `POST /v1/experiments/quantum_volume` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"qubits": [0, 1], "trials": 5, "shots": 128, "seed": 7}
response = httpx.post(f"{base_url}/experiments/quantum_volume", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
