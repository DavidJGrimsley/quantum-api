# Randomized Benchmarking
Calls `POST /v1/experiments/randomized_benchmarking` with `QUANTUM_API_BASE_URL` and `QUANTUM_API_KEY`.

```python
import os

import httpx

base_url = os.environ["QUANTUM_API_BASE_URL"].rstrip("/")
api_key = os.environ["QUANTUM_API_KEY"]
payload = {"qubits": [0], "sequence_lengths": [1, 2, 4, 8], "num_samples": 3, "shots": 256, "seed": 7}
response = httpx.post(f"{base_url}/experiments/randomized_benchmarking", json=payload, headers={"X-API-Key": api_key}, timeout=30.0)
response.raise_for_status()
response.json()
```
